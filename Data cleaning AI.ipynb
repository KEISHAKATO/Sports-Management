```python
from google.colab import files
import zipfile
import os

# Upload the file
uploaded = files.upload()

# Get the uploaded file's name
zip_file_name = next(iter(uploaded))

# Define extraction directory
extract_dir = '/content/extracted_files'

# Unzipping the file
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# Listing the extracted files
extracted_files = os.listdir(extract_dir)
extracted_files
```

```python
import pandas as pd

# Path to the extracted CSV file
csv_file_path = '/content/extracted_files/Airplane_Crashes_and_Fatalities_Since_1908.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
fatality_locations = df[['Date', 'Location', 'Aboard', 'Fatalities']]
fatality_locations.head()
```

```python
max_fatalities_date = fatality_locations.loc[fatality_locations['Fatalities'].idxmax()]['Date']
print(f"The date with the highest fatalities is: {max_fatalities_date}")
```

```python
no_fatalities = fatality_locations[fatality_locations['Fatalities'] == 0]
print(f"Number of crashes with no fatalities: {len(no_fatalities)}")
```

```python
# Split the 'Location' column into two parts: 'Region' and 'Country/State'
split_location = fatality_locations['Location'].str.split(',', n=1, expand=True)

# Create new columns for 'Region' and 'Country/State', handling cases where there's no comma
fatality_locations['Region'] = split_location[0]
fatality_locations['Country/State'] = split_location[1].fillna('Unknown')

# Display the first few rows
fatality_locations.head()
```

```python
top_100_fatalities = fatality_locations.sort_values(by='Fatalities', ascending=False).head(100)
top_100_fatalities.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_25_fatalities = fatality_locations.groupby('Country/State')['Fatalities'].sum().nlargest(25)
plt.figure(figsize=(10, 8))
top_25_fatalities.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top 25 Fatalities by Country/U.S. State')
plt.ylabel('')
plt.show()
```
